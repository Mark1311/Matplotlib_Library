```python
# Import_Library

import matplotlib.pyplot as plt
import numpy as np
```

# Bar Chart

```python
X = ["A", "B", "C", "D", "E"]
Y = [25,65,95,80,20]

plt.xlabel("Alphabt", fontsize = 20)
plt.ylabel("% Values", fontsize = 15)
plt.title("This is A Bar Chart")
plt.bar(X,Y, width = 0.5, edgecolor = "red", label = "Height of green", linestyle = ":", alpha = 0.99 )
plt.show()
```

# Scatter Plot

```python
X = ["A", "B", "C", "D", "E","F","G","H","I","J","K"]
Y = [25,65,75,80,20,40,45,55,10,35,85]

plt.xlabel("Alphabt", fontsize = 20)
plt.ylabel("Values", fontsize = 15)
plt.title("This is A Scatter Plot")
plt.scatter(X,Y, color ="Red", alpha = 0.99, marker = "+", linewidth = 2)
plt.show()
```

# Histogram Plot

```python
Data  = np.random.randn(100)
plt.title("Histogram Chart")
plt.xlabel("Alphabt", fontsize = 15)
plt.ylabel("Values", fontsize = 15)

plt.hist(Data, bins = 20, edgecolor = 'black')
plt.show()
```

# Pie Chart

```python
Data = [23,50,46,89,56]
Y = ["A","B","C","D","E"]
ex = [0.3,0.0,0.2,0.0,0.0]
plt.title("Pie Chart", fontsize = 20)
plt.pie(Data, labels = Y, explode = ex, shadow = True,hatch=['**O', 'oO', 'O.O', '.||.'])
plt.legend(Y, loc = 1)
plt.axis('equal')
plt.show()

# Others Atteributes
#Radies = 1.75
# label distance = 1.2
# taxtprops= {"fontsize":18}
# CounterClock = True/False
# Rotate Labels = True/False
# wedgeprops = {"linewidth" : 1.3, "edgecolor" : "Green"}
# startangle = 45
```

# Dot Chart

```python
Data = [130]
plt.pie(Data, labels = ['One Value'], shadow = True, radius=0.7, hatch='**O')
plt.legend(labels = ['Dot'])
plt.title("Dot Chart", fontsize = 20)
plt.show()
```

# Double Pie Chart

```python
Data1 = [20,20,20,20]
Data2 = [50,50,50,50]
ex1 = [0.1,0.1,0.1,0.1]
ex2 = [0.1,0.1,0.1,0.1]

plt.pie(Data1, radius=1, explode = ex1, hatch=['**O', 'oO', 'O.O', '.||.'])
plt.pie(Data2, radius = 0.65, explode = ex2, startangle=45)
plt.title("OverLaping Pie Chart")
plt.show()
```

# Stem Plot

```python
x = np.arange(4, 12, 1)
y = x ** 2

plt.stem(x,y, markerfmt = 'D', linefmt= ':', basefmt = "red", bottom = 1.1)
plt.title("Stem Plot", fontsize = 20)
plt.show()

#Another Arguments
# use_line_collection = True/False
```

# Box Plot

```python
Data = [12,20,40,200,30,90,50,65,100]
plt.boxplot(Data, vert = False, widths=0.25, showmeans=True,sym = "gd")
plt.title("Box Plot", fontsize = 20)
plt.show()

#Atteributes
# notch = True/False
```

# Stack Or Area-Stack Plot

```python
Data = [1,2,3,4,5]
area1= [5,15,10,25,45]
area2 = [10,55,65,75,45]
area3 = [65,60,50,70,55]
label = ("area1","area2","area3")

plt.stackplot(Data,area1,area2,area3, baseline = "zero")
plt.title("Stack Area Plot", fontsize = 20)
plt.legend(label, loc = 2)
plt.show()
```

# Step Plot

```python
Datax = [1,2,3,4,5,6]
Datay = [2,4,6,8,10,12]

plt.step(Datax, Datay, marker = "D", ms = 5, mfc = "green")
plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
plt.title("Step Plot", fontsize = 20)
plt.show()
```

# Fill Between Plot

```python
Datax = [1,2,3,4,5]
Datay = [1,2,3,4,15]
plt.fill_between(Datax, Datay, color= "Black")
plt.title("Fill Between Plot", fontsize = 20)
plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
plt.show()
```

```python
plt.plot(Datax, Datay, color = "Green")
plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
plt.title("Fill Between Plot", fontsize = 20)
plt.show()
```

# Sub Plot

```python
Datax = [1,2,3,4,5]
Datay = [2,4,6,8,10]

# Syntax = subplot(RoW, Colm, Index)

# Chart- 1
plt.subplot(2,2,1)
plt.plot(Datax,Datay)
plt.title("Line Plot", fontsize = 15)

# Chart- 2
plt.subplot(2,2,2)
plt.pie(Datax)
plt.title("Pie Plot", fontsize = 15)
# Chart- 3
plt.subplot(2,2,3)
plt.pie(Datay)
plt.title("Pie Plot", fontsize = 15)

# Chart- 4
plt.subplot(2,2,4)
plt.fill_between(Datax, Datay, color= "Black")
plt.title("Fill Between Plot", fontsize = 15)


plt.show()
```

# How to write text on the graph?

```python
# Create a line plot
Data = [0, 12,32,35,34,65,61,37,58,49]
plt.plot(Data)

# Syntax = (x, y, "Your Value", styel here)
plt.text(2,8, "My Text", style = "normal")
plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
plt.title("Line Plot")
plt.figure(figsize = (2,2))
plt.show()
```
